```python
from matplotlib import pyplot as plt
import time
```

```python
with open("histogram.log",'r') as file_handle:
    file_data = file_handle.read()
```

```python
!head -n 4 histogram.log
```

```python
results = {}
for line in file_data.split("\n"):
    count_and_size = list(filter(None,line.strip().split(" ")))
    if len(count_and_size)>1:
        results[int(count_and_size[1])] = int(count_and_size[0])
```

```python
plt.scatter(x=results.keys(),y=results.values())
plt.xlabel('file size [bytes]', fontsize=16)
plt.ylabel('number of files with that size', fontsize=16)
ax = plt.gca()
ax.set_yscale('log')
ax.set_xscale('log')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('file_sizes.png')
```
